```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
# Open-Meteo API URL (Pune location example)
url = "https://api.open-meteo.com/v1/forecast?latitude=18.52&longitude=73.85&hourly=temperature_2m"

response = requests.get(url)

data = response.json()

# Extract temperature and time
time = data['hourly']['time']
temperature = data['hourly']['temperature_2m']

print("Data fetched successfully!")
df = pd.DataFrame({
    'Time': time,
    'Temperature (C)': temperature
})

df.head()
plt.figure(figsize=(10,5))
plt.plot(df['Time'][:24], df['Temperature (C)'][:24])
plt.xticks(rotation=45)
plt.xlabel("Time")
plt.ylabel("Temperature (C)")
plt.title("Hourly Temperature for Pune")
plt.tight_layout()
plt.show()
import seaborn as sns

sns.set(style="whitegrid")

plt.figure(figsize=(10,5))
sns.lineplot(x=df['Time'][:24], y=df['Temperature (C)'][:24])
plt.xticks(rotation=45)
plt.title("Hourly Temperature Visualization")
plt.show()



```
