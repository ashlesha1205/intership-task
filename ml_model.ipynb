```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

# Step 2: Create sample dataset
data = {
    'text': [
        'Win money now',
        'Call me today',
        'Limited time offer',
        'Meeting at 5pm',
        'Congratulations you won prize',
        'Let us have lunch'
    ],
    'label': [
        'spam',
        'ham',
        'spam',
        'ham',
        'spam',
        'ham'
    ]
}

df = pd.DataFrame(data)

# Step 3: Convert labels to numbers
df['label'] = df['label'].map({'ham':0, 'spam':1})

# Step 4: Split data
X_train, X_test, y_train, y_test = train_test_split(
    df['text'], df['label'], test_size=0.3, random_state=42
)

# Step 5: Convert text to numerical data
vectorizer = CountVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# Step 6: Train model
model = MultinomialNB()
model.fit(X_train_vec, y_train)

# Step 7: Predict
y_pred = model.predict(X_test_vec)

# Step 8: Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```

This project implements a spam email classification model using Scikit-Learn.
The model uses CountVectorizer for text feature extraction and Multinomial Naive Bayes for classification.
The dataset is split into training and testing sets.
The model is evaluated using accuracy and classification report.
